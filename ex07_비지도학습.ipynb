## iris 데이터 
- 차원 축소

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X=iris.data
y=iris.target
```

```python
X
```

### PCA
- 주성분 분석을 사용해서 4차원 데이터를 2차원으로 축소

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) # 전체 데이터를 2차원으로 축소
```

```python
pca.fit(X)
```

```python
X_trans=pca.transform(X)
X_trans
```

```python
# 축소된 데이터 시각화
plt.figure(figsize=(10,10))
# 산점도
plt.scatter(X_trans[:,0], # 축소된 0번 컬럼
            X_trans[:,1], # 축소된 1번 컬럼
            c = y) 
plt.show()
```

```python
from sklearn.datasets import fetch_lfw_people
```

```python
X = people['data']
y = people['target']
```

```python
# 픽셀값 = 0 ~ 255
X[0]
```

```python
# 픽셀값 범위를 0 ~ 1
X = X / 255.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
# knn모델 학습
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
knn.score(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
pca = PCA(n_components=100, whiten=True)
```

```python
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
knn.fit(X_train_pca, y_train)
```

```python
knn.score(X_train_pca, y_train)
```

```python
knn.score(X_test_pca, y_test)
```

```python
plt.imshow(pca.components_[0].reshape(87,65), cmap='gray')
plt.show()
```

## 2차원 데이터 셋

```python
from sklearn.datasets import make_blobs # 2차원 데이터셋 생성
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(random_state=2)
```

```python
X.shape
```

```python
y
```

```python
# 군집 모델 생성
kmeans =KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
pred= kmeans.predict(X)
```

```python
pred
```

```python
# 중심 포인트의 좌표
```

```python
kmeans.cluster_centers_
```

```python
for n_cluster in range(3):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
# Kmeans 군집을 사용해서 나눈 그룹 시각화
center = kmeans.cluster_centers_
for n_cluster in range(3):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.scatter(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

```python
# KMeans
## 모든 방향의 중요도가 같다고 생각함
## 데이터가 특정한 모양을 가지고 이쓴 경우 그룹을 잘 만들지 못함
```

## 반달 데이터셋

```python
from sklearn.datasets import make_moons
X, y =make_moons(n_samples=200, random_state=0, noise=0.05)
```

```python
for n_cluster in range(2):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
pred=kmeans.predict(X)
```

```python
center = kmeans.cluster_centers_
for n_cluster in range(2):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

## 병합군집

```python
mglearn.plots.plot_agglomerative_algorithm()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg =AgglomerativeClustering(n_clusters=2)
pred=agg.fit_predict(X)
```

```python
for n_cluster in range(2):
  data = X[pred == n_cluster]
  plt.scatter(data[: , 0] , data[: , 1] , label = n_cluster)

plt.legend()
plt.show()
```

## DBSCAN
- 밀도 기반 클러스터링
- Density Based Spatial Clustering of Applications with noise

```python
mglearn.plots.plot_dbscan()
```

```python
# DBSCAN은 스케일리을 진행하는 것이 좋음
```

```python
from sklearn.preprocessing import StandardScaler
st_scaler = StandardScaler()
st_scaler.fit(X)
X_scaler = st_scaler.transform(X)
```

```python
from sklearn.cluster import DBSCAN
dbscan =DBSCAN(eps=0.5, min_samples=5)
pred=dbscan.fit_predict(X_scaler)
```

```python
pred
```

```python
# 데이터들이 모여있는 방향을 알 수 있음
for n_cluster in range(2):
  data = X[pred == n_cluster]
  plt.scatter(data[: , 0] , data[: , 1] , label = n_cluster)

plt.legend()
plt.show()
```
